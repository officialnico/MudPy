```python
from mud import Player, World, MUDIndexerSDK
from cafecosmos import *
```

```python
abi_dir = "./cafecosmos-contracts/abi"
rpc = "https://rpc.garnetchain.com"
world_address = "0x0209FE2b6C1a2B8cE12D8d9D6cf2D78a3Ff324FE"

indexer_url = "https://indexer.mud.garnetchain.com/q"
mud_config_path = "./cafecosmos-contracts/mud.config.ts"

cafecosmos = World(rpc, 
              world_address, 
              abi_dir, 
              indexer_url=indexer_url, 
              mud_config_path=mud_config_path)

```

```python
player = Player(env_key_name="PLAYER1")
player.add_world(cafecosmos, "cafecosmos")
```

```python
landId = find_player_lands(cafecosmos, player.player_address, 1)[0]
```

```python
from web3 import Web3

function_call = player.cafecosmos.placeItem(landId, 5, 5, name_to_id_fuzzy("pink floor"), mode="raw")

# Estimate the gas required for the transaction
estimated_gas = function_call.estimate_gas({
    "from": player.player_address,
})

# Get the current gas price from the network
current_gas_price = cafecosmos.w3.eth.gas_price

# Build the transaction with estimated gas and gas price
txn = function_call.build_transaction({
    "chainId": cafecosmos.chain_id,  # Ethereum Mainnet
    "gas": estimated_gas,
    "gasPrice": current_gas_price,
    "nonce": cafecosmos.w3.eth.get_transaction_count(player.player_address),
})

# Sign the transaction
signed_txn = cafecosmos.w3.eth.account.sign_transaction(txn, player.private_key)
```

```python
# Send the transaction
txn_hash = cafecosmos.w3.eth.send_raw_transaction(signed_txn.raw_transaction)
print(f"Transaction sent. TX hash: {txn_hash.hex()}")

# Wait for receipt
txn_receipt = cafecosmos.w3.eth.wait_for_transaction_receipt(txn_hash)
print("Transaction confirmed:", txn_receipt)
```
